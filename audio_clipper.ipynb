```python
from python_speech_features import fbank
from python_speech_features import logfbank
import scipy.io.wavfile as wav
import glob
from scipy.io.wavfile import read
import numpy as np
from pydub import AudioSegment
from pydub.utils import make_chunks

for filename in glob.glob('Hindi dataset/*.wav'):
    
    myaudio = AudioSegment.from_file(filename , "wav") 
    chunk_length_ms = 60000 # pydub calculates in millisec
    chunks = make_chunks(myaudio, chunk_length_ms) #Make chunks of one minute
    
    #Export all of the individual chunks as wav files
    audio_name = filename.replace("Hindi dataset\\","")
    audio_name = audio_name.replace(".wav","")
    print(audio_name)
    
    for i, chunk in enumerate(chunks):
        chunk_name = ("clipped hindi audio/" + audio_name + "chunk{0}.wav").format(i)
        print("exporting", chunk_name)
        chunk.export(chunk_name, format="wav")
    
```
